# Traditional Programming vs Neural Network

In traditional programming, the rules and data are given and then the answers are obtained. For example, given a list of x values, the program outputs the corresponiding y values using the rules defined in function:

```
def function(x):
  y = []
  for num in x:
    y.append(2*num -1)
  return y
```

```
x = [-1.0,  0.0, 1.0, 2.0, 3.0, 4.0]
y = function(x)
print (y)
```

Let's train a neural network using TensorFlow to do the equivalent task. Using the x and y data, we train a neural network to figure out the relationship between them:

# 1. Define and compile the *network*

```
import numpy as np
import tensorflow as tf
from tensorflow import keras

# the model has only 1 layer with 1 neuran
model = tf.keras.models.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

# 2. Providing the data

```
x = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
y = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

# 3. Training the neural Network

```
model.fit(x, y, epochs=500)
```

# 4. Check how well the neural network can predict unseen x data

```
model.predict([10])
```
